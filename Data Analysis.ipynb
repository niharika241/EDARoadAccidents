## Univariate Analysis

### 

```python
#Gender of Driver
sns.countplot(x=df['Sex_of_driver'])
```

```python
#Severity of Accident
x = ((df['Accident_severity'].value_counts()/len(df))*100)
plt.pie(x.values,labels=x.index,autopct='%1.1f%%')
plt.show()
```

```python
# Distribution of Number of Accidents in Age Groups
d = df['Age_band_of_driver'].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(x=d.values,y=d.index,orient = 'h')
plt.ylabel('Age Group')
plt.xlabel('Number of Accidents')
plt.title('Number of Accidents by Age Group')
plt.show()
```

```python
# Accidents by hour 
df1=df
df1['Hour']=pd.to_datetime(df1['Time']).dt.hour
plt.figure(figsize=(12,4))
sns.countplot(x=df1['Hour'])
plt.ylabel('Number of Accidents')
plt.show()
```

```python
d = df['Cause_of_accident'].value_counts()
plt.figure(figsize=(18,10))
squarify.plot(sizes = d.values, label = d.index, color = sns.color_palette("pastel", len(d.values)),ec='black')
plt.axis("off")
plt.show()
```
